# MovieLens Statistics Analysis

This notebook shows distributions of user and item statistics from the MovieLens data.

This notebook does not compute those statistics itself â€” first run the `rating-stats.py` script to produce the `item-stats.csv` and `user-stats.csv` this notebook requires.

This repository is not limited to analyzing the `ml-latest-small` data set.  If you work with a different version of the data, such as ML-25M, change the following variable to change where the notebook looks for data:

```python
ML_DIR = 'ml-latest-small'
```

## Setup

Let's import some Python modules:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## User Analysis

Let's look at user data first.

```python
user_stats = pd.read_csv(f'{ML_DIR}/user-stats.csv')
user_stats.head()
```

Let's plot the distribution of rating counts:

```python
sns.displot(user_stats['count'])
plt.show()
```

And of average ratings:

```python
sns.displot(user_stats['mean'])
plt.show()
```

The user average rating looks somewhat left-skewed.

## Item Stats

Let's do a similar analysis with the item stats.

```python
item_stats = pd.read_csv(f'{ML_DIR}/item-stats.csv')
item_stats.head()
```

Item rating count distribution:

```python
sns.displot(item_stats['count'])
plt.show()
```

That is very skewed. Let's look at a log-plot CDF:

```python
sns.ecdfplot(item_stats['count'], log_scale=10)
plt.xlabel('# of Ratings')
plt.ylabel('Cum. Frac. Movies')
plt.show()
```

And we can look at the item average ratings:

```python
sns.displot(item_stats['mean'])
plt.show()
```
